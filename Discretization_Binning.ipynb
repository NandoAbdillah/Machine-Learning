<h2> Binning dengan ketentuan sendiri</h2>

```python
from sklearn.pipeline import Pipeline
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('./main/datasetgaji.csv')
data.head()
```

```python
#Membuat batas manual
batas_bin = [0,1400000,4000000] #variabel batas bin penggambarannya -> 0 > n < 1400000 > n > 4000000
```

```python
kategori = ['kecil', 'besar']
```

```python
data['gaji_binned_1'] = pd.cut(data['Gaji Orang Tua'], bins=batas_bin, labels=kategori)
```

```python
print(data)
```

<h2>Binning dengan linespace</h2>

```python
bins = np.linspace(min(data['Gaji Orang Tua']), max(data['Gaji Orang Tua']), 3)
# jadi batasnya adalah dengan mengambil nilai minimum & maksimum , lalu 
# mesin akan menetapkan 1 batas tengah dengan jarak dari minimum ke batas tengah , dan batas tengah ke maksimum sama
# nilai 3 karena banyak data ada 6 sehingga jika ingin ada 2 kategorikal maka harus diabagi 3
```

```python
bins
```

```python
kategori = ['kecil', 'besar']
data['gaji_binned2'] = pd.cut(data['Gaji Orang Tua'], bins=bins, labels= kategori, include_lowest = True)
#include_lowest apabila true itu berarti apakah kita akan memasukkan data yang paling rendah atau tidak
```

```python
data
```

<h2>Binning dengan Quantile</h2>

```python
data['gaji_binned3'] = pd.qcut(data['Gaji Orang Tua'], 2)
data
```

```python
data['gaji_binned3'] = pd.qcut(data['Gaji Orang Tua'], q=2, labels=kategori)
data
```
