<h2> Data Normalization Min-Max </h2>

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('./main/Data Normalization and Discretization/datasetgaji2.csv')
```

```python
df
```

```python
scaler = MinMaxScaler() # Lihat rumus min max di ss
```

```python
scaled = scaler.fit_transform(df[['Gaji Orang Tua', 'Umur']])
# Mengubah gaji orang tua dan umur
```

```python
scaled
#kiri gaji orang tua, kanan umur
#data yang paling tinggi contoh disitu adalah gajinya 4 juta
# maka x maks gaji 4000000 adalah 1
# dan x maks umur 28 adalah 1
# sedangkan data yang paling sedikit diwakili dengan 0 seperti dibawah
# gaji orangtua dyrga dan umur rangga sasana
```

```python
# cara baca data diatas adalah dengan mengangap itu adalah persen
# contoh 0.07 adalah sekitar 7% -> maka dibaca data ini adalah 7% dari nilai tertinggi
# rentang skala adalah 0-1
```
