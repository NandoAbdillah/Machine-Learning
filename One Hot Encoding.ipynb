```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
```

```python
dict = {
    'ID' : np.arange(11101,11107),
    'Nama Produk' : ['Tomat', 'Apel', 'Pisang', 'Semangka', 'Melon', 'Jeruk'],
}
np.arange(11101,11107)
```

```python
data = pd.DataFrame(dict)
data
```

```python
# Mengubah tipe data numerik menjadi kategorik
cat = OneHotEncoder() #sparse digunakan untuk mengubah kategorik menjadi array seperti hasil dibawah ini
cat = cat.fit_transform(data[['Nama Produk']])
cat = cat.toarray()
cat
#Transformasi binner dibawah diurukan berdasarkan abjad nama produk berarti apel adaalah kolom pertama dan kolom terakhir adalah tomat
```

```python
data_cat = pd.DataFrame(cat)
data_cat
```

```python
header = data['Nama Produk'].sort_values()
data_cat.columns = 'Produk_' + header
data_cat
```

```python
data_join = data.join(data_cat)
data_join
```
