<h2>1. Mengimport Library untuk EDA (Exploratory Data Analysis) </h2>

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

<h2>2. Loading Dataset </h2>

```python
df = pd.read_csv('./main/data_car.csv')
df.head()
```

<h2>3. Mengecek Tipe Dari Data </h2>

```python
df.dtypes
```

<h2> 4. Mengubah Nama Kolom </h2>

```python
df.rename(columns= {'Engine HP': 'HP', 'Engine Cylinders' : 'Cylinders', 'Transmission Type' : 'Transmission', 'Driven_Wheels' : 'Drive Mode', 'highway MPG' : 'MPG-H', 'MSRP' : 'price'}, inplace = True) 
```

```python
df
```

<h2> 5. Dropping Kolom </h2>

```python
df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'],axis = 1, inplace = True)
```

```python
df
```

<h2> 6. Drop Baris Duplikat </h2>

```python
df.shape
```

```python
duplicated_rows_df = df[df.duplicated()] #Untuk memeriksa baris yang duplikat
print('number of duplicated rows : ' , duplicated_rows_df.shape)
```

<h2> 9.Plot yang berbeda fitur terhadap satu sama lain (scatter) terhadap frekuensi(Histogram)  </h2>
#Cara menyikapi outliers
#Diskritisasi mengkelompokkan contoh harga -> mengkelompokkan sesuai dengan murah dan mahal
#Normalisasi dengan cara diskala 

```python
df.price.describe().apply(lambda x : format(x, '.5f'))
#std = standard deviasi
```

```python
# Analisisi Unvariate
# df.Make.value_counts().plot(kind = 'bar', figsize=(10,5))
df.Make.value_counts().nlargest(40).plot(kind = 'bar', figsize=(10,5))
# bisa dikasih nlargest untuk menampilkan 40 sja 
plt.title('Number of cars by make')
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
# Heat Map
# Analisis Bivariate
plt.figure(figsize=(10,5))
c = df.corr(numeric_only = True) # Agar nilai numerik saja yang ditampilkan jika tidak akan error
sns.heatmap(c, cmap = 'RdYlGn' , annot= True)
c
```

```python
#Analisis Multivariate

result = pd.pivot_table(data= df, index= 'Transmission', columns='Drive Mode', values='price')
print(result)

sns.heatmap(result, annot=True, cmap ='RdYlGn', center = 0.117)
plt.show()
```

```python
# Scatter Plot  -> Digunakan untuk bivariate
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['price'])
ax.set_xlabel('HP')
ax.set_ylabel('price')
plt.show()
```

```python
sns.pairplot(data = df, vars=['HP', 'Drive Mode', 'price'])
plt.show()
```

```python
# Univariate -> analisis satu kolom dengan value_count nya
# Bivariate -> analisis fitur/kolom dengan target  (2 )
# Multivariate -> analisis fitur/kolom dengan fitur/kolom dengan target (3/lebih)
```
