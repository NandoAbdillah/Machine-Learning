```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv('./main/Supervised Learning/golf.txt')
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

# Inisialisasi OneHotEncoder
encoder = OneHotEncoder()

# Fit dan transform data
one_hot = encoder.fit_transform(df[['Outlook', 'Temp.','Humidity','Wind']])

# Konversi sparse matrix ke array numpy
encode = one_hot.toarray().astype(int)

# Membuat DataFrame baru dengan nama kolom yang dihasilkan oleh OneHotEncoder
df_encode = pd.DataFrame(encode, columns = encoder.get_feature_names_out())

```

```python
df_encode
```

```python
## atau bisa menggunakan cara dibawah ini yang lebih simpel

x = df.iloc[:, :-1]
y = df.iloc[:, 4]
```

```python
x.values
```

```python
one_hot = pd.get_dummies(x, drop_first = False).astype(int)
one_hot.head()
#Fungsi pd.get_dummies digunakan untuk mengubah data kategori menjadi angka 0 dan 1, proses ini disebut one-hot encoding. Misalnya, jika kita memiliki kolom ‘Warna’ dengan nilai ‘Merah’, ‘Biru’, dan ‘Hijau’, maka get_dummies akan membuat tiga kolom baru: ‘Warna_Merah’, ‘Warna_Biru’, dan ‘Warna_Hijau’. Setiap baris data akan memiliki nilai 1 pada kolom yang sesuai dengan warnanya, dan 0 pada kolom lainnya.
#Parameter drop_first digunakan untuk menghindari masalah yang disebut multicollinearity. Jika drop_first = True, maka kolom pertama (dalam contoh kita, ‘Warna_Merah’) akan dihapus. Alasannya adalah, jika kita tahu bahwa ‘Warna_Biru’ dan ‘Warna_Hijau’ adalah 0, maka kita pasti tahu bahwa ‘Warna_Merah’ harus 1. Jadi, kita bisa menghapus ‘Warna_Merah’ tanpa kehilangan informasi apa pun.
#Namun, dalam kasus Anda, drop_first = False, yang berarti semua kolom (Warna_Merah, Warna_Biru, dan Warna_Hijau) akan dipertahankan. Ini bisa berguna jika Anda ingin melihat semua informasi dalam data Anda.
```

```python
df = df.drop(x, axis =1)
df = pd.concat([one_hot,df], axis =1)
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
X,y
```

```python
# Datalatih (train) harus lebih banyak dari data uji (test)
# information_gain , data_entrophy yang dihitung hanyalah dari data latih

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
#dapat diketahui bahwa data test kita set 30% dari banyak data sdngkan data train adalah 70% dari banyak data
#fungsi dari random_state adlah python akan mengambil data acak 30% dan 70% dari jumlah data yang ada, jadi tidak dalam urutan data
#jika tanpa random state maka ketika kita run berulang kali maka data train dan test akan acak terus berubah ubahm sehingga permodelannya juga akan berubah ubah
```

```python
#melihat banyak elemen pada setiap data train dan data test
result = list(map(len,[X_train, X_test, y_train, y_test]))
print(result)
```

```python
model = DecisionTreeClassifier(criterion= 'entropy', random_state =100)
#jika criterion = 'gini' , digunakan untuk pencarian root/akar
model.fit(X_train, y_train)
```

```python
# model validation atau memprediksi, kan disini udah difit di x_train, dan y_train, kemudian mendapat pola prediksi
# lalu di y_pred ini kita uji data yang masi sisa X_test untuk divalidasi atau memprediksi target sesuai dengan pola prediksi di awal
# mencari tahu hasilnya sudah baik atau belum
# model validation/prediction
y_pred = model.predict(X_test)
```

```python
print('Accuracy: ' , metrics.accuracy_score(y_test, y_pred))
```
